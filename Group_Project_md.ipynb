### plot line graphs

```python
def plot_it(country_name: str):
    data = time_df_t[country_name]
    plot_line = plt.plot(data, label="Cases", color="yellowgreen", marker='o')
    plt.figsize=(10, 3)
    plt.xticks([])
    plt.title(country_name)
    plt.xlabel('Day(1/22 - 3/14)')
    plt.ylabel('number of cases')
    plt.tight_layout()
    plt.show()
```

```python
country_names = time_df_t.columns
```
